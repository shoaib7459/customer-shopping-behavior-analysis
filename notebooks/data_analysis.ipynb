## Handling Missing Values in Review Ratings

### Objective
To address the 37 missing values in the review_rating column using category-based imputation for accurate and context-aware data completion.

### Approach
Using median imputation grouped by product category to:
- Maintain category-specific rating patterns
- Preserve data distribution
- Avoid bias from overall dataset median
- Handle outliers appropriately

### Code Implementation

```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

- **Why median?** Robust to outliers, preserves central tendency

## Outlier Detection in Purchase Amounts

### Objective
To identify potential outliers in purchase amounts using the Interquartile Range (IQR) method, ensuring data quality and detecting any unusual transaction patterns.

### Methodology: IQR Rule
- **Q1 (25th percentile):** First quartile
- **Q3 (75th percentile):** Third quartile  
- **IQR:** Q3 - Q1 (spread of middle 50% of data)
- **Lower Bound:** Q1 - 1.5 × IQR (capped at minimum 0)
- **Upper Bound:** Q3 + 1.5 × IQR
- **Outliers:** Values falling outside these bounds

### Code Implementation

```python
Q1 = df['purchase_amount'].quantile(0.25)
Q3 = df['purchase_amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
lower_bound = max(0, lower_bound)
upper_bound = Q3 + 1.5 * IQR
lower_bound, upper_bound
```

### 📊 Outlier Analysis Results

**IQR Calculation:**

| Metric | Value |
|--------|-------|
| **Q1 (25th percentile)** | $39.00 |
| **Q3 (75th percentile)** | $81.00 |
| **IQR (Q3 - Q1)** | $42.00 |
| **Lower Bound** | $0.00 (adjusted from negative) |
| **Upper Bound** | $144.00 |

**Outlier Boundaries:**
- Acceptable range: **$0.00 to $144.00**
- All purchase amounts: **$20.00 to $100.00**

### ✅ Outlier Detection Outcome


```python
outliers = df[
    (df['purchase_amount'] < lower_bound) | df['purchase_amount'] > upper_bound
]
outliers.shape
```
